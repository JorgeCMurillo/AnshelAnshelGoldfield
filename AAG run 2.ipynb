Then, given $G$, we will compute $g_i^{-1}$, where if g = $x_1 \cdot x_2 \cdot  ... \cdot x_m$, $g^{-1} = x_m^{-1}x_{m-1}^{-1} \cdot ... \cdot x_2^{-1} \cdot x_1^{-1}$.

```python
def inverse_list(Group_G):
    # assuming our group G consists of g_i, 1<=i<= 20.
    # we can interpret G as a matrix/array. we construct a temporary array that will be our G^-1
    temp_array = []
    #we focus on each row of our "matrix"/ array G , since each row correspondes to g_i, i in 1,2,..20
    for gi in Group_G:
        list_index = len(gi)
        # we construct a temporary row / list
        temp_list = []
        for i in range(list_index):
            # we switch the order of the elements in each row / list,  and negate each term. I.e the first term goes to last term, 
            #so we get g_i^-1 = g^{-1} = x_m^{-1}x_{m-1}^{-1} \cdot ... \cdot x_2^{-1} \cdot x_1^{-1}
            temp_list.append(-1*gi[list_index - i-1])
        # we return G^-1
        temp_array.append(temp_list)        
    return temp_array
```

```python
def inverse_array(Group_G): 
    # we  reverse the rows, switching the first row with last, second row with second to last, etc.
    #create empty array
    temp_array = []
    #determine how many rows there are
    length = len(Group_G)
    #here we make the last row of group G the first of temp_array, second to last row the second row of temp array, etc
    for i in range(length):
        temp_array.append(Group_G[length-1-i])
    return temp_array
```

we now define the public conjugate. suppose we have alice's key, we want to send publicly the conjugates  ag_1a^-1,ag_2a^-1,...,ag_na^-1. so we define the public_conj function to do this.

```python
def fuse(array):
    #here we fuse any array of lists into 1 list
    list = []
    for i in range(len(array)):
        list += array[i]
    return list
```

```python
def public_conj(x,g):
#    we want to compute xg_1x^-1,...,xg_nx^-1
    x_k = fuse(x)
    x_k_i = fuse(inverse_array(inverse_list(x)))
    temp_array = []
    temp_list = []   
    for i in range(len(g)):
        temp_list = x_k + g[i]+ x_k_i
        temp_array.append(temp_list)

    return temp_array
public_conj(a,G)
```

we also need to be able to compute $xg_1^{-1}x^{-1},...,xg_m^{-1}x^{-1},$ and hence define the public_conj_neg function to do this

```python
def public_conj_neg(x,g):
#    we want to compute xg_1^{-1}x^{-1},...,xg_m^{-1}x^{-1},
    #with x being an array of g_i,...,g_k
    # we fuse x to become g_i*...*g_k
    x_k = fuse(x)
    # we fuse x^-1 to become g_k^-1*...*g_i
    x_k_i = fuse(inverse_array(inverse_list(x)))
    temp_array = []
    temp_list = []    
    for i in range(len(g)):
        # we compute x(g_1^-1)x^-1,...,xg_m^-1x^-1
        temp_list = x_k + fuse(list(inverse_list([g[i]])))+ x_k_i
        temp_array.append(temp_list)
    return temp_array
```

```python
def return_indices(x,g):
    #given a private key x, we return it's indices i,k,l of it's decomposition g_i*g_k*g_l
    indices =[]
    for i in range(len(x)):
        indices.append(g.index(x[i]))
    return indices 
```

if we want x to communicate with y, we construct a secret key, which given x  , uses publicly available conjugates 

yg_1y^-1,...yg_ny^-1 

and 

yg_1^-1y^-1,...yg_m^-1y^-1 

to compute both x(yx^-1y^-1) and  yxy^-1x^-1. 
If alice, denoted as a uses the secret key to communicate with bob, and bob does as well, they will together
compute

aba^-1b^-1

and 

bab^-1a^-1.
we check to see if they share one key in common

```python
def secret_key(x,y,g):
    keys = []
    
    # we will compute two keys, key1 will compute yxy^-1x^-1, key2 will compute xyx^-1y^-1.
    
    # if y is an array of the form g_i,...,g_k
    # 
    # then y^-1 = g_k^-1,...,g_1^-1
    # we can compute xy^-1x^-1   by finding the indices of g_k,...,g_i in the public_conj_neg function
    # and computing xg_k^-1x^-1,...,xg_ix^-1
    y_i_indices = return_indices(inverse_array(y),g)
    public_conjugate = []
    # we create an array containing the conjugates xg_k^-1x^-1,...,xg_ix^-1
    # which we later fuse to get xy^-1x^-1
    for i in range(len(y_i_indices)):
        public_conjugate.append(public_conj_neg(x,g)[y_i_indices[i]])
    y_i = inverse_array(inverse_list(y))
    # we compute y(xy^-1x^-1) by fusing arrays to get y * xy^-1x^-1
    key_1 = [] + fuse(y)+ fuse(public_conjugate)   
    keys.append(B_80(key_1))
    
    # we can compute xyx^-1   by finding the indices of g_i,...,g_k in the public_conj function
    y_indices = return_indices(y,g)
    public_conjugate = []
    for i in range(len(y_i_indices)):
        # we construct an array containing xg_ix^-1,...,xg_kx^-1
        public_conjugate.append(public_conj(x,g)[y_indices[i]])
    # here we compute xyx^-1x^-1
    key2 = [] + fuse(public_conjugate) + fuse(y_i)
    keys.append(B_80(key2))
    return keys
```
